# TQuant Lab 安裝教學

## 使用的套件以及環境
* Main package: Zipline-reloaded 2.2.0
* Python 3.8
* Microsoft Windows OS or macOS
* 其他在執行 Zipline-TEJ 時需要用到的套件: Pandas, Numpy, Logbook, Exchange-calendars

## 安裝 Zipline-tej

* 在開始安裝之前，我們建議使用 [Anaconda](https://www.anaconda.com/data-science-platform) 作為開發環境。

* 需要下載以下安裝包。

    1. Windows [(zipline-tej.yml)](https://minhaskamal.github.io/DownGit/#/home?url=https://github.com/tejtw/zipline-tej/blob/main/zipline-tej.yml)

    2. Mac [(zipline-tej_mac.yml)](https://minhaskamal.github.io/DownGit/#/home?url=https://github.com/tejtw/zipline-tej/blob/main/zipline-tej_mac.yml)


* 接著，在 Anaconda 的命令提示符中，創建一個虛擬環境並安裝相應版本的依賴包。我們強烈推薦使用虛擬環境，以保持每個項目的獨立性。[(原因說明)](https://csguide.cs.princeton.edu/software/virtualenv#definition)

```
    Window 使用者

    # 將下載好的 zipline-tej.yml檔案放在以下路徑中
    $ cd <C:\Users\username\Downloads>

    # 創建虛擬環境
    $ conda env create -f zipline-tej.yml

    # 開啟虛擬環境
    $ conda activate zipline-tej

    Mac 使用者
    
    # 將下載好的 zipline-tej.yml檔案放在以下路徑中
    $ cd <C:\Users\username\Downloads>

    # 創建虛擬環境
    $ conda env create -f zipline-tej_mac.yml

    # 開啟虛擬環境
    $ conda activate zipline-tej
```

* 除了使用 YML 文件外，如果你對 Python 相當熟悉，也可以手動創建虛擬環境，並逐一安裝所需的依賴包。

```
    # create virtual env
    $ conda create -n <env_name> python=3.8

    # activate virtual env
    $ conda activate <env_name>

    # download dependency packages
    $ conda install -c conda-forge -y ta-lib
    $ conda install -y notebook=6.4.11
    $ conda install -c conda-forge nb_conda_kernels
    $ conda install -y xlrd=2.0.1
    $ conda install -y openpyxl=3.0.9
    $ pip install zipline-tej

```


* 注意，我們需要先下載TA-lib，這樣我們才能成功下載zipline-tej。

## 解決交易日曆問題

由於 TQuant Lab 使用 TEJ 所提供的台灣證券交易資料，所以我們需要使用專屬於台股的交易日曆，以便於後續回測使用。下方提供了可供下載的交易日曆文件，需要將這些文件放置在指定的  **exchange_calendars** 文件夾中。  [download](https://minhaskamal.github.io/DownGit/#/home?url=https://github.com/tejtw/zipline-tej/tree/main/exchange_calendars)

在下載了上面提到的交易日曆文件後將 **exchange_calendar_tejxtai.py** 複製，並將該檔案添加到 **exchange_calendars** 文件夾中的 **calendar_utils.py** 中。

導航至 site packages 中的 exchange_calendars 文件夾。通常它位於以下路徑：

    Window 使用者

    **C:\Users\username\Anaconda3\envs\zipline-tej\Lib\site-packages\exchange_calendars**

    Mac 使用者

    **C:\Users\username\Anaconda3\envs\zipline-tej\Lib\python3.8\site-packages\exchange_calendars**


但是，對於一些Windows用戶，也可能位於以下路徑：

    **C:\Users\username\\AppData\Roaming\Python\Python38\Scripts**

如果發生這種情況，建議將 **exchange-calendars** 文件夾放置在我們上面提到的前一個路徑中。
這樣做的目的是將新的交易日曆文件放置在正確的位置，以便 **Zipline-tej** 能夠使用自定義的交易日曆進行回測。
